```python
import sys
import pathlib
sys.path.append(str(pathlib.Path().resolve().parent))

#-- Basics Library
import numpy as np
import pandas as pd
import os 
from sklearn.preprocessing import MinMaxScaler

#-- Warnings OFF
from warnings import simplefilter
import joblib

from utils import vec_2_esc, df_transform
from narmax import *

simplefilter("ignore", FutureWarning)
np.seterr(all="ignore")
os.environ['KERAS_BACKEND'] = 'theano'
os.environ['PYTHONHASHSEED']=str('123')


%matplotlib inline
```

```python
#-- Parameters
look_back = 12
num_month = 12
vec_num = 12

#-- Training and fitting
test_size = 12*10
epochs=1000
batch_size= 120
validation_split=0.1


RAW_PATH = '../Datasets/Raw/'
PROCESSED_PATH = '../Datasets/Processed/'
RESULTS_PATH = '../Results/Models/'
FINAL_PATH = '../Datasets/Final/'
FIGURE_PATH = '../Figure/Results/'
```

```python
def preprocess_data(file_name, name,RAW_PATH,PROCESSED_PATH,save_data=True,save_scaler=True):
    df_raw = df_transform(f'{RAW_PATH}/{file_name}',
                        name)

    scaler = MinMaxScaler()
    df_raw['value_norm'] = scaler.fit_transform(df_raw[['value']])
    df_raw = df_raw[['date','name','value','value_norm']]

    if save_data:
        df_raw.to_csv(f'{PROCESSED_PATH}/{name}.csv',index=False)

    if save_scaler:
        joblib.dump(scaler, f'{RESULTS_PATH}/{name}_scaler.pkl')

    return df_raw
```
